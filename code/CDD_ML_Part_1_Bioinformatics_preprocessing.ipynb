### What is ChEMBL Database

‘chemogenomic’ database that brings together chemical, bioactivity and genomic data to aid the translation of genomic information into effective new drugs. The resource is therefore of interest to drug discovery researchers in large pharmaceutical companies, as well as small companies and academic institutions.

```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

Target: Replicase polyprotein 1ab affects organism: Severe acute respiratory syndrome coronavirus 2

```python
selected_target = targets.target_chembl_id[10]
selected_target
```

Here, we will retrieve only bioactivity data reported as IC50 values in nM unit

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head()
```

Finally we will save the resulting bioactivity data to a CSV file bioactivity_data.csv.

```python
df.to_csv('bioactivity_data_raw.csv', index=False)
# Saving the dataframe named df containing the bioactivity data into a new csv file
# Ensure dataframe's index is not written as a column in the csv file
```

```python
df2 = df[df.standard_value.notna()]
# filtering out null or missing standard_value columns so only valid standard_value data remain
```

filtered out around 144 rows

## Data preprocessing of bioactive data

lets label compounds as either active, inactive or intermediate by binning the standard_value measured in nM. Standard_value are essentially like the concentration of compound required to modulate the target protein and cause a biological effect.

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
   bioactivity_class.append("intermediate")
```

```python
len(bioactivity_class)
```

## Combine 3 columns(molecule_chembl_id, canonical_smiles, standard_value) and bioactivity_class into a dataframe

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2[selection]
df3[df3.standard_value.notna()]
```

```python
df4 = df3.reset_index(drop=True)
df4
```

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

```python
# Save dataframe to csv file
df5.to_csv('bioactivity_data_preprocessed.csv', index=False)
```
